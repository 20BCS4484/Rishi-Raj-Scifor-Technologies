```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data
np.random.seed(0)
data = pd.DataFrame({
    'A': np.random.normal(0, 1, 100),
    'B': np.random.normal(1, 1, 100),
    'C': np.random.normal(2, 2, 100),
    'Category': np.random.choice(['X', 'Y', 'Z'], 100)
})
```

```python
sns.pointplot(data=data, x='Category', y='C')
plt.title('Point Plot')
plt.show()
```
