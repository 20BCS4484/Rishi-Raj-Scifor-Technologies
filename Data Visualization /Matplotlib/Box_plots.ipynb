```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/IRIS.csv'  # Replace with your actual file path
iris_df = pd.read_csv(file_path)

```

```python
numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
species = iris_df['species'].unique()

```

```python
for column in numerical_columns:
    plt.figure(figsize=(6, 4))  # Smaller figure size
    iris_df.boxplot(column=column, by='species', grid=False)
    plt.title(f'Box Plot of {column} by Species')
    plt.suptitle('')  # Suppress the default title to avoid duplication
    plt.xlabel('Species')
    plt.ylabel(column)
    plt.show()
```
