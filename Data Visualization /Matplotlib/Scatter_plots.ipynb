```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/IRIS.csv'  # Replace with your actual file path
iris_df = pd.read_csv(file_path)

```

```python
numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
species = iris_df['species'].unique()

```

```python
# Plot scatter plots for each pair of numerical columns
for i in range(len(numerical_columns)):
    for j in range(i + 1, len(numerical_columns)):
        plt.figure(figsize=(6, 4))
        for specie in species:
            subset = iris_df[iris_df['species'] == specie]
            plt.scatter(subset[numerical_columns[i]], subset[numerical_columns[j]], label=specie)
        plt.title(f'Scatter Plot of {numerical_columns[i]} vs {numerical_columns[j]}')
        plt.xlabel(numerical_columns[i])
        plt.ylabel(numerical_columns[j])
        plt.legend(title='Species')
        plt.show()
```
