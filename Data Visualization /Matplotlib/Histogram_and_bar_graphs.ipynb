```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/IRIS.csv'  # Replace with your actual file path
iris_df = pd.read_csv(file_path)

```

```python
# Define numerical columns
numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Plotting bar graphs for sum of numerical values grouped by species
for column in numerical_columns:
    species_sum = iris_df.groupby('species')[column].sum()
    species_sum.plot(kind='bar', title=f'Sum of {column} by Species', xlabel='Species', ylabel=f'Sum of {column}', color=['blue', 'orange', 'green'])
    plt.show()

```

```python
# Plotting histograms for each numerical column
for column in numerical_columns:
    plt.hist(iris_df[column], bins=10, edgecolor='black', color='purple')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Bar plots for the average values of each numerical column grouped by species
for column in numerical_columns:
    species_avg = iris_df.groupby('species')[column].mean()
    species_avg.plot(kind='bar', title=f'Average of {column} by Species', xlabel='Species', ylabel=f'Average of {column}', color=['blue', 'orange', 'green'])
    plt.show()

```

```python
# Bar plots for the maximum values of each numerical column grouped by species
for column in numerical_columns:
    species_max = iris_df.groupby('species')[column].max()
    species_max.plot(kind='bar', title=f'Maximum of {column} by Species', xlabel='Species', ylabel=f'Maximum of {column}', color=['blue', 'orange', 'green'])
    plt.show()
```
